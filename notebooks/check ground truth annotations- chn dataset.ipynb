```python
from detectron2.data import MetadataCatalog, DatasetCatalog    
import random
import os
import matplotlib.pyplot as plt
from detectron2.data.datasets import register_coco_instances
```

```python
import pickle
import json
import cv2 
from detectron2.utils.visualizer import Visualizer
from detectron2.data.detection_utils import read_image
```

```python
images_path = '/data/cddod/doc_chs_median_release/JPEGImages/'
json_path = '/data/cddod/doc_chs_median_release/coco_converted/cddod_coco.json'
name = 'chn'
```

```python
register_coco_instances(name, {}, json_path, images_path)
```

```python
MetadataCatalog.get("dla_val").thing_classes = ['text', 'title', 'list', 'table', 'figure']
```

```python
MetadataCatalog.get(name)
```

```python
a = DatasetCatalog.get(name)
```

```python
for d in random.sample(a, 1):
    img = cv2.imread(os.path.join(images_path, d["file_name"]))
    visualizer = Visualizer(img[:, :, ::-1], 
                            metadata=MetadataCatalog.get(name), scale=5)
    out = visualizer.draw_dataset_dict(d)
    plt.figure(figsize=(20,20))
    plt.imshow(out.get_image()[:, :, ::-1])
    plt.show()
```
